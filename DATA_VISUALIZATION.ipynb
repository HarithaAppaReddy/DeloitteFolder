```python
from matplotlib import style
style.use('ggplot')
x=[5,8,10]
y=[12,16,6]
x1=[6,9,11]
y1=[6,15,7]
plt.plot(x,y,'g:',label='Line one',linewidth=1,marker='o')
plt.plot(x1,y1,'r-.',label='Line two',linewidth=3)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Graph")
plt.grid(True, color='k')
plt.legend()
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g:',label='Line one',linewidth=1,marker='o')
plt.xlabel("X1-Axis")
plt.ylabel("Y1-Axis")
plt.title("GRAPH-1")
plt.legend()
plt.subplot(1,2,2)
plt.plot(x1,y1,'r-.',label='Line two',linewidth=3)
plt.xlabel("X2-Axis")
plt.ylabel("Y2-Axis")
plt.title("GRAPH-2")
plt.legend()
plt.suptitle("graph")
```

```python
#seaborn
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("tips.csv")
df.head()
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'],data=df,estimator=np.std)
```

```python
sns.countplot(x=df['sex'],data=df)
```

```python
#boxplot
sns.boxplot(x=df['sex'],y=df['total_bill'],data=df)
```

```python
sns.boxplot(x=df['tip'],y=df['total_bill'],data=df)
#sns.boxplot(x=df['tip'])
```

```python
sns.boxplot(data=df, palette='rainbow')
```

```python
sns.boxplot(x=df['day'],y=df['total_bill'],hue="sex",data=df,palette="coolwarm")
```

```python
sns.swarmplot(x="total_bill",y="day",data=df)
```

```python
sns.violinplot(x=df["sex"],y=df["total_bill"],data=df)
```

```python
#strip plot
sns.stripplot(x="day",y="total_bill",data=df)
```

```python
sns.stripplot(x="day",y="total_bill",data=df,jitter=True,hue='sex',palette='Set1')
```

```python
sns.displot(df["total_bill"])
```

```python
sns.displot(df["tip"],kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='purple')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```
